# HeatMap

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars=pd.read_csv('./mpg.csv')
```

```python
cars_origin = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
cars_origin
```

```python
sns.heatmap(cars_origin)
```

```python
sns.heatmap(cars_origin.T, cmap = 'YlGn')
```

```python
a = np.array([[1,2],
              [3,4]
             
             ])
```

```python
sns.heatmap(a)
```

```python
sns.heatmap(cars.corr())
```

```python
sns.heatmap(cars.corr(), vmin= -4, vmax=3, center=4)
```

```python
cars.corr()
```

```python
sns.heatmap(cars.corr(), 
            cmap='gist_earth_r', 
            center=0, 
            vmin=0, vmax=1, 
            annot = True,
            linewidth = 3
           
           )
```

```python

sns.heatmap(cars.corr(), 
            cmap='tab20c', 
            center=0, 
            vmin=0, vmax=1, 
            annot = True,
            linewidth = 3,
            square= False
           
           )
plt.savefig('./fig25.png',dpi= 300,bbox_inches = 'tight')
```
