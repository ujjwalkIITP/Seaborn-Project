# Box Plot, Violin Plot, Swarm plot and stripplot

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('./mpg.csv')
```

```python
cars.head()
cars.dropna(inplace = True)
```

```python
cars.describe()
```

```python
cars.head()
```

# Graph Box Plot

```python
sns.boxplot(x='mpg', data = cars)
plt.show()
```

```python
sns.boxplot(y='mpg', data = cars, x = 'origin')
```

```python
sns.boxplot(y='mpg', data = cars, x = 'origin', hue = 'cylinders')
plt.savefig('./fig5.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
cars['cylinders'].value_counts()
```

```python
cars = cars[cars['cylinders'].isin([4,6,8])]
```

```python
cars['cylinders'].value_counts()
```

```python
sns.boxplot(y='mpg', data = cars, 
            x = 'origin', 
            hue = 'cylinders',
           order = ['europe','japan','usa'])
plt.show()
```

```python
cars['model_year'].describe()
```

```python
cars['newer_model'] = cars['model_year']> 76
```

```python
sns.boxplot(x = 'origin',
           y = 'mpg',
           hue = 'newer_model',
            data = cars,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
           )
plt.plot()
```

```python
cars1=cars[cars['cylinders']==4]
```

```python
sns.boxplot(x = 'origin',
           y = 'mpg',
           hue = 'newer_model',
            data = cars1,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
           )
plt.plot()
```

```python
cars2=cars[cars['cylinders']==6]
```

```python
sns.boxplot(x = 'origin',
           y = 'mpg',
           hue = 'newer_model',
            data = cars2,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
           )
plt.plot()
```

# VIOLIN PLOT

```python
sns.violinplot(x='mpg', data = cars)
```

```python
sns.violinplot(y='mpg',
               x = 'origin',
               data = cars,
               hue = 'cylinders'
              
              )
```

```python
sns.violinplot(x = 'origin',
           y = 'mpg',
           hue = 'newer_model',
            data = cars1,
            split = True,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
           )
plt.plot()
```

```python
cars3=cars[cars['cylinders'].isin([4,6])]
```

```python
sns.violinplot(x = 'origin',
           y = 'mpg',
           hue = 'cylinders',
            data = cars3,
            split = True,
            scale = 'count',
            order=['japan', 'europe', 'usa'],
           )
plt.plot()
```

```python
sns.violinplot(x = 'origin',
           y = 'mpg',
           hue = 'cylinders',
            data = cars3,
            split = True,
            scale = 'count',
            order=['japan', 'europe', 'usa'],
            inner = 'quartiles'
           )
plt.plot()
plt.savefig('./fig4.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```

# SWARM PLOT

```python
a = [4,4,4,4,1,1,1,1,2,2,5,5,7]
a.sort()
sns.swarmplot(a, 
              color='blue', 
              size=15, 
              linewidth=3, 
              edgecolor='yellow'
             )
plt.tight_layout()
```

```python
sns.swarmplot(x = 'mpg', y = 'origin', data = cars)
```

```python
USA = cars[cars['origin']=='usa']
sns.violinplot(y = 'mpg', x = 'cylinders', data = USA,inner = None,  scale = 'width', color = 'black' )
sns.swarmplot(y = 'mpg', x = 'cylinders', data = USA, alpha = 1, dodge = True, color = 'red', size = 3)
#sns.violinplot(x = 'mpg', y = 'origin', data = cars, hue = 'cylinders')
plt.savefig('./fig1.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
sns.violinplot(x=USA.cylinders, y=USA.horsepower,
               scale='width', inner=None, color='xkcd:forest green'
              )
sns.swarmplot(x=USA.cylinders, y=USA.horsepower,
              color='white', size=3
             );
plt.savefig('./fig2.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
USA = cars[cars['origin']=='usa']
sns.boxplot(y = 'mpg', x = 'cylinders', data = USA,  color = 'black' )
sns.swarmplot(y = 'mpg', x = 'cylinders', data = USA, alpha = 1, dodge = True, color = 'yellow')
#sns.violinplot(x = 'mpg', y = 'origin', data = cars, hue = 'cylinders')
```

# Stripplot

```python
a = [1,1,1,1,2,2,2,5,5,7,7,7]
sns.stripplot(a, size = 20, alpha = .8, linewidth = 3, jitter = True)
plt.show()
```

```python
sns.stripplot(x = 'cylinders', y = 'mpg', data = USA, linewidth = 2, size = 7)
```

```python
sns.stripplot(x = 'origin', y = 'weight', data = cars)
```

```python
sns.set_style('white')
sns.stripplot(cars.weight,['']*len(cars), hue=cars.cylinders, 
              palette=sns.cubehelix_palette(3));
sns.despine(left=True)
plt.xlabel('\nWeight (lbs)')
plt.legend(title='Cylinders', frameon=False);
plt.savefig('./fig3.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```
