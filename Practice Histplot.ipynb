# Visualization Begins

```python
sns.set_style('white')
sns.histplot(x='bill_length_mm', data = penguins)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True)
#plt.show()
plt.savefig('./fig11.jpeg', dpi = 200, bbox_inches = 'tight')
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, bins = 20)
#plt.show()
jupyter labextension install jupyterlab-plotly
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, bins = [0,10,20,40,50,60,100])
plt.xticks(range(0,100,10))
plt.show()
```

```python
b=[]
for i in range(9):
    b.append(i*i + 2*i)
print(b)
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, bins = b)
plt.xticks(b)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, binwidth=10, binrange=(30,60))
plt.tight_layout()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, binwidth=5, binrange=(30,60),
            stat = 'count')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, binwidth=5, binrange=(30,60),
            stat = 'probability')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, binwidth=5, binrange=(30,60),
            stat = 'probability', fill = False, element = 'step', cumulative = True)
plt.tight_layout()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(x='bill_length_mm', data = penguins, kde = True, binwidth=5, binrange=(30,60),
            stat = 'density')
plt.tight_layout()
plt.show()
```

```python
sns.histplot(x='bill_length_mm', data = penguins, hue = 'species', fill = False, element = 'bars')
plt.show()
```

```python
sns.histplot(x='bill_length_mm', data = penguins, hue = 'species', fill = True, element = 'poly')
#plt.show()
plt.savefig('./fig13.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
sns.histplot(x='bill_length_mm', data = penguins, 
             hue = 'species', 
             fill = True, 
             element = 'bars', 
             multiple = 'stack')
#plt.show()
plt.savefig('./fig7.png', dpi = 500, bbox_inches = 'tight', transparent = True)
```

```python
sns.histplot(x='bill_length_mm', data = penguins, 
             hue = 'species', 
             fill = True, 
             element = 'bars',
             
             multiple = 'fill')
plt.show()
```

```python
sns.histplot(x='bill_length_mm', data = penguins, 
             hue = 'species', 
             fill = True, 
             element = 'bars', 
             multiple = 'dodge')
plt.show()
```

```python
sns.histplot(x='bill_length_mm', data = penguins, 
             hue = 'species', 
             fill = False, 
             element = 'bars',
             shrink = 0.8,
             multiple = 'layer')
plt.show()
```

```python
sns.histplot(x='bill_length_mm', 
             y='bill_depth_mm', 
             data = penguins, 
             legend=True,
             palette='bone',
             cbar=True
             )
plt.show()
```

```python
sns.histplot(x='bill_length_mm', 
             y='species',
             data = penguins, 
             hue = 'species',
             )
#plt.show()
plt.savefig('./fig8.png', dpi = 500, bbox_inches = 'tight', transparent = True)
```

```python
sns.histplot(x='species', 
             y='island',
             data = penguins, 
             hue = 'species'           
             )
plt.show()
```
