```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
diamond = pd.read_csv('./diamonds.csv')
```

```python
diamond.head()
```

```python
sns.scatterplot(x='cut',y='table' ,hue = 'table',linewidth = 1, data=diamond)
plt.plot()
plt.legend([],[], frameon=False)
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='cut',y='table' ,hue = 'table',linewidth = 1, data=diamond, style = 'cut', s= 100)
plt.plot()
plt.legend([],[], frameon=False)
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='cut',y='table' ,
                hue = 'table',linewidth = 1, s = 500,
                data=diamond, style = 'cut')
plt.plot()
plt.legend([],[], frameon=False)
plt.savefig('./fig24.png',dpi= 300,bbox_inches = 'tight')
```
